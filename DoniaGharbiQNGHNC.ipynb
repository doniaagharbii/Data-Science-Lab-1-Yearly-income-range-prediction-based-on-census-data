##  Pre-processing

### Encoding

```python
# Identify Numeric features
numeric = ['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week','income']

# Identify Categorical features
catategorical = ['workplace','education','marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
```

```python
df['income']=df['income'].map({'<=50K': 0, '>50K': 1}).astype(int)
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
dataset = df.copy()

for x in df[catategorical].columns:
    dataset[x] = LabelEncoder().fit_transform(df[catategorical][x])
dataset.head()
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

train_col_sacle = dataset[['age','fnlwgt','education-num','hours-per-week','capital-gain','capital-loss']]
train_scaler_col = scaler.fit_transform(train_col_sacle)
train_scaler_col = pd.DataFrame(train_scaler_col,columns=train_col_sacle.columns)

dataset['age']= train_scaler_col['age']
dataset['fnlwgt']= train_scaler_col['fnlwgt']
dataset['education-num']= train_scaler_col['education-num']
dataset['hours-per-week']= train_scaler_col['hours-per-week']
dataset['capital-gain']= train_scaler_col['capital-gain']
dataset['capital-loss']= train_scaler_col['capital-loss']
dataset.head()
```

### Feature selection

```python
from scipy.stats import chi2_contingency, ttest_ind

def contingency_table(feature):
    p = pd.crosstab(dataset['income'],dataset[feature])
    return(p)
feature = []
pval = []
chi2_result = []

for i in catategorical[:]:
    feature.append(i)
    result = chi2_contingency(contingency_table(i))
    pval.append(round(float(result[1]),6))
    
    if float(result[1]) < 0.01:
        chi2_result.append("Significant")
    else:
        chi2_result.append("Insignificant")
```

```python
chisquare = pd.DataFrame(data={'PValue':pval,'Result':chi2_result},index=feature)
chisquare
```

```python
def t_test(df,columns):    
    feature = []
    pval = []
    ttest_result = []
    
    for i in columns:
        feature.append(i)
        result = ttest_ind(dataset[dataset["income"]==0][i], dataset[dataset["income"]==1][i])
        pval.append(round(float(result[1]),6))
        
        if float(result[1]) < 0.01:
            ttest_result.append("Significant")
        else:
            ttest_result.append("Insignificant")
    
    return(pd.DataFrame(index=feature,data={'PValue':pval,'Result':ttest_result}))
```

```python
norm_ttest = t_test(dataset, dataset[numeric])
norm_ttest
```

```python
dataset = dataset.drop('fnlwgt',1)
dataset.head()
```

# Classification

### Splitting data

```python
X = dataset.drop(['income'],1)
y = dataset['income']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42, stratify=y)
```

```python
print(X_train.shape, y_train.shape,X_test.shape, y_test.shape)
```

### Naive Bayes

#### Without GridSearch

```python
from sklearn.naive_bayes import GaussianNB


# instantiate the model
gnb = GaussianNB()
# fit the model
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score :", str(round(accuracy_score(y_test, y_pred)*100, 2)),'%')

from sklearn.metrics import f1_score
print("F1 Score :",str(round(f1_score(y_test, y_pred, average='weighted')*100, 2)),'%')
```

#### With GridSearch

```python
from sklearn.model_selection import GridSearchCV,cross_val_score, KFold
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_roc_curve,recall_score
import warnings
```

```python
def grid_search(X = X_train,y = y_train):
    
    warnings.filterwarnings(action='ignore',message='')
    scoring ='accuracy'  
    param_grid_nb = {
    'var_smoothing': np.logspace(0,-9, num=100)
    }
   
    score=0
    GS = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid_nb, verbose=1, cv=10, n_jobs=-1)
    grid_result = GS.fit(X, y)
    print (f'best accuracy score of GaussianNB:' + str(round(grid_result.best_score_*100, 2)) + f' % \n best estimator: {grid_result.best_estimator_} \n best params: {grid_result.best_params_}\n')
        
    trained_model_NB=grid_result.best_estimator_.fit(X,y)
    y_predic = trained_model_NB.predict(X)
    print(classification_report(y,y_predic)+'\n')
```

```python
trained_model_NB=grid_search(X_train,y_train)
```

### Logistic Regression

#### Without GridSearch

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
#Train our model with the training data
LR.fit(X_train, y_train)
#print our price predictions on our training data
y_pred = LR.predict(X_train)
```

```python
print("Accuracy score :",str(round(accuracy_score(y_train, y_pred)*100, 2)),'%')
print("F1 Score :",str(round(f1_score(y_train, y_pred, average='weighted')*100, 2)),'%')
```

#### With GridSearch

```python
def grid_search(X = X_train,y = y_train):
    
    warnings.filterwarnings(action='ignore',message='')
    scoring ='accuracy'  
    param_grid_lr = { 'C':[0.01,0.1,1,10,100], 'max_iter':[100,1000,2500],'penalty':['l1','l2'], 'solver':["liblinear","saga"]}
   
    score=0
    GS = GridSearchCV(estimator=LogisticRegression(), param_grid=param_grid_lr, verbose=1, cv=10, n_jobs=-1)
    grid_result = GS.fit(X, y)
    print (f'best accuracy score of LogisticRegression:' + str(round(grid_result.best_score_*100, 2)) + f' % \n best estimator: {grid_result.best_estimator_} \n best params: {grid_result.best_params_}\n')
        
    trained_model_LR=grid_result.best_estimator_.fit(X,y)
    y_predic = trained_model_LR.predict(X)
    print(classification_report(y,y_predic)+'\n')
```

```python
trained_model_LR=grid_search(X_train,y_train)
```

## Clustering

### KMeans

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters=3 , init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X)
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(dataset, kmeansmodel.labels_, metric='euclidean')

print('Silhouetter Score: %.3f' % score)
```

```python
from sklearn.decomposition import PCA

labels_kmeans=kmeansmodel.labels_

pca = PCA(n_components=2)
res = pca.fit_transform(X)

plt.scatter(res[labels_kmeans==0,0], res[labels_kmeans==0,1],s=50,label='0',alpha=0.8)
plt.scatter(res[labels_kmeans==1,0], res[labels_kmeans==1,1],s=50,label='1',alpha=0.8)
plt.scatter(res[labels_kmeans==2,0], res[labels_kmeans==2,1],s=50,label='2',alpha=0.8)

plt.title('Kmeans clustering')
plt.show()
```

# Frequent pattern mining

```python
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
catf = df.select_dtypes(include=['object']).columns
dfcat = df[catf].copy()
```

```python
dfcp = pd.get_dummies(dfcat)
dfcp_apr = apriori(dfcp, min_support = 0.6, use_colnames = True, verbose = 1)
dfcp_apr2 = association_rules(dfcp_apr, metric = "confidence", min_threshold = 0.6)
```
